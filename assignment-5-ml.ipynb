Create Different between order now from the previous one

```python
size = cust_order_buy3.shape[0]
#temp = pd.DataFrame(columns=['A'])
a = []
for i in range (size-1) :
    if cust_order_buy3.iloc[i,0] == cust_order_buy3.iloc[i+1,0] :
        a.append((cust_order_buy3.iloc[i+1,1] - cust_order_buy3.iloc[i,1])/np.timedelta64(1,'M'))
    else :
        a.append((cust_order_buy3.iloc[i,1] - cust_order_buy3.iloc[i,1])/np.timedelta64(1,'M'))
```

```python
diff = pd.DataFrame(a, columns = ["diff"])
diff = round(diff)
diff.head()
```

```python
diff.dtypes
```

```python
cust_order_buy4 = pd.concat([cust_order_buy3, diff], axis=1)
cust_order_buy4.tail(10)
```

```python
cust_order_buy4.isna().sum()
```

```python
cust_order_buy4["diff"] = cust_order_buy4["diff"].fillna(0)
cust_order_buy4
```

```python
def flags(df) : 
    if df["diff"] <= 0 : 
        return 1
    elif df["diff"] <= 3 :
        return 0
    else :
        return 1

    
cust_order_buy4["flagging"] = cust_order_buy4.apply(flags, axis =1 )
```

```python
def nonevent(df) : 
    return (df[0]/937)
def event(df) : 
    return (df[1]/95144)
def Woe(df) :
    return ln(df["%nonevent"]/df["%event"])
```

```python
a["%nonevent"] = a.apply(nonevent, axis =1 )
a["%event"] = a.apply(event, axis =1 )

b["%nonevent"] = b.apply(nonevent, axis =1 )
b["%event"] = b.apply(event, axis =1 )

c["%nonevent"] = c.apply(nonevent, axis =1 )
c["%event"] = c.apply(event, axis =1 )

d["%nonevent"] = d.apply(nonevent, axis =1 )
d["%event"] = d.apply(event, axis =1 )
```

Split Data

```python
x = ['review_scoremax','Category_payment','geolocation_state']
y = ['flagging']
from sklearn.model_selection import train_test_split
x=result3[x]
y=result3[y]
x_train, x_test, y_train, y_test = train_test_split(x,y ,train_size=0.8)
print("Banyaknya data trainning=", len(x_train))
print("Banyaknya data testing=", len (x_test))
```

Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
clf=MultinomialNB()
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
confusion_matrix(y_test, y_pred) 
print (classification_report(y_test, y_pred))
```

DTree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
dt = DecisionTreeClassifier (max_depth=None, min_samples_split=2)
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
skor_dt= metrics.accuracy_score(y_test, y_pred)
print( "akurasi klasifikasi dengan D tree adalah=",skor_dt)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
# Predicting the Test set results
y_pred = classifier.predict(x_test)
confusion_matrix(y_test, y_pred) 
print (classification_report(y_test, y_pred))
```
